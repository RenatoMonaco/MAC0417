```python
from skimage import filters

from skimage import morphology

os.chdir(gray_folder)

gray_images = os.listdir()

for img_name in gray_images:

  os.chdir(gray_folder)

  img = image.imread(img_name)
  img_denoised = filters.median(img, selem=np.ones((7,7)))
  img_canny = feature.canny(img_denoised, sigma=2)
  img_canny = morphology.remove_small_objects(img_canny, min_size=5)

  closed = morphology.binary_dilation(img_canny, selem=np.ones((15,15)))
  clean1 = morphology.remove_small_objects(closed, min_size=500)
  clean1 = morphology.binary_closing(clean1, selem=np.ones((15,15)))
  clean1 = morphology.convex_hull_object(clean1)
  clean = morphology.remove_small_objects(clean1, min_size=20000)
  segmentada = morphology.convex_hull_image(clean)

  os.chdir(segmentada_folder)

  imsave(img_name, img_as_ubyte(segmentada))




# # display results
# fig, (ax1, ax2, ax3, ax4, ax5, ax6) = plt.subplots(nrows=1, ncols=6, figsize=(60, 10),
#                                     sharex=True, sharey=True)

# ax1.imshow(img, cmap=plt.cm.gray)
# ax1.axis('off')
# ax1.set_title('cinza', fontsize=20)

# ax2.imshow(segmentada, cmap=plt.cm.gray)
# ax2.axis('off')
# ax2.set_title('segmentada', fontsize=20)

# ax3.imshow(img_canny, cmap=plt.cm.gray)
# ax3.axis('off')
# ax3.set_title('canny', fontsize=20)

# ax4.imshow(clean, cmap=plt.cm.gray)
# ax4.axis('off')
# ax4.set_title('clean', fontsize=20)

# ax5.imshow(closed, cmap=plt.cm.gray)
# ax5.axis('off')
# ax5.set_title('dilated', fontsize=20)

# ax6.imshow(clean1, cmap=plt.cm.gray)
# ax6.axis('off')
# ax6.set_title('clean1', fontsize=20)




# plt.show()
```
